## Subtract the highest RGB value from the image

```python
rgb_max_value = 255
reduced_img = rgb_max_value - img
```

## Show the new image 

```python
plt.imshow(reduced_img)
```

## Flip the new Image

```python
plt.imshow(np.flip(reduced_img))
```

```python
print(" Notice how the flip function changed the color of the image as well")
print(" This is due to the transpose of the image entry values")
```

## Rotate the image by 90 degrees

## Create a black and white Image 

```python
print(" This can only be done with the aid of the gray- val formular")
print(" Also divide the image by 255")
```

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
simg = img/255
print(simg)
```

```python
print(" Multiply both the resulting division and constant gray-value")
```

```python
sRGB = np.matmul(simg, grey_vals)
```

```python
plt.imshow(sRGB)
```

```python
print(" Show the graph and equate the cmap to gray")
```

```python
plt.imshow(sRGB, cmap='gray')
```
